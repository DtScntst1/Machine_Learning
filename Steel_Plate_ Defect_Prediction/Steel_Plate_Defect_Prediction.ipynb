```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
# Importing the data
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")

# Checking the number of rows and columns of data
print("train: ",train.shape)
print("test: ", test.shape)
```

```python
# Create input properties by dropping id and 'Pastry', 'Z_Scratch', 'K_Scatch', 'Stains', 'Dirtiness', 'Bumps', 'Other_Faults' columns

X = train.drop(['id','Pastry', 'Z_Scratch', 'K_Scatch', 'Stains', 'Dirtiness', 'Bumps', 'Other_Faults'], axis=1)
y = train[['Pastry','Z_Scratch', 'K_Scatch', 'Stains', 'Dirtiness', 'Bumps', 'Other_Faults']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=0)
```

```python
# Create an XGBoost Classifier using the XGBClassifier class and then fit the training data to the model using the fit() method
from xgboost import XGBClassifier

classifier = XGBClassifier()

classifier.fit(X_train, y_train)
```

```python
# Make predictions based on input features (X_test) in the test set using a trained regression model
y_pred = classifier.predict(X_test)
```

```python
print(y_pred)
```

```python
# Create a new data frame by removing certain column from the data frame named "test"
test1 = test.drop(columns = 'id')
```

```python
# Create a NumPy array by selecting all rows and columns of dataframe test1
unseen = test1.iloc[:, :].values
```

```python
# Make predictions on "unseen" data with a model called classifier
unseen_pred = classifier.predict(unseen)
```

```python
import pandas as pd

# Creating DataFrame using 'unseen_pred' array
pred_df = pd.DataFrame(unseen_pred, columns=['Pastry','Z_Scratch', 'K_Scatch', 'Stains', 'Dirtiness', 'Bumps', 'Other_Faults'])

# Add new columns to your 'test' dataframe
test = pd.concat([test, pred_df], axis=1)
```

```python
# Train a classifier for each label sequentially and then calculate the AUC score by making predictions on the test set
from sklearn.metrics import roc_auc_score
for i in y:
    classifier.fit(X_train, y_train[i])
    ctbproba = classifier.predict_proba(X_test)[:, 1] 
    print(f'AUC Score - {i}: {roc_auc_score(y_test[i],ctbproba)}\n')
    pred = classifier.predict_proba(X_test)[:, 1]
```

```python
#  Create a DataFrame containing predictions
submission = test[['id', 'Pastry','Z_Scratch', 'K_Scatch', 'Stains', 'Dirtiness', 'Bumps', 'Other_Faults']]

#  Convert to the format required by the submission file (for example, CSV format)
submission.to_csv('submission.csv', index=False)
```
